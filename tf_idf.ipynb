```python
from pathlib import Path
import csv
import pandas as pd
```

```python
test = pd.read_csv('equalized_train.csv')
```

```python
test
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(1, 5))
vectors = vectorizer.fit_transform(test['0'].values.astype('U'))
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 5))
vectors = vectorizer.fit_transform(test['0'].values.astype('U'))
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
dff = pd.DataFrame(denselist, columns=feature_names)
```

```python
tdf = pd.concat([df,dff], axis = 1)
```

```python
tdf
```

```python
add_label = test.iloc[:,-1:]
```

```python
add_label
```

```python
tdf = pd.concat([tdf,add_label], axis=1)
```

```python
tdf.to_csv("tf-idf_train.csv")
```
