```python
import pandas as pd
import numpy as np
from statistics import mean
import csv
from sklearn import svm
from pathlib import Path
import sklearn.metrics as metric
from sklearn.model_selection import KFold
import time
```

```python
start = time.time()
#read data in chunks of 1 million rows at a time
chunk = pd.read_csv('tf-idf_train_first_svm.csv',chunksize=10)
end = time.time()
print("Read csv with chunks: ",(end-start),"sec")
pd_df = pd.concat(chunk)
```

```python
data = pd_df.iloc[:,1:]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

header = list(data.columns)
header.pop()
cc = pd.Categorical(data['1']).codes
X_train, X_test, y_train, y_test = train_test_split( data[header], cc, test_size=0.2, random_state=0)
```

```python
cc
```

```python
svc = svm.SVC(kernel='poly', C = 0.1).fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
result = metric.accuracy_score(y_test, y_pred)
print(result)
result1 = metric.f1_score(y_test, y_pred, average= 'macro')
print(result1)
result3 = metric.recall_score(y_test, y_pred, average = "macro")
print(result3)
result2 = metric.precision_score(y_test, y_pred, average = "macro")
print(result2)
```

```python
kf = KFold(n_splits=10, random_state=None)
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
ac = []
f1 = []
re = []
pr = []
```

```python
for train_index , test_index in kf.split(X):
    X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train , y_test = y[train_index] , y[test_index]
     
    svc = svm.SVC(kernel='poly', C = 0.1).fit(X_train,y_train)
    pred_values = svc.predict(X_test)
    result = metric.accuracy_score(y_test, pred_values)
    ac.append(result)
    print(result)
    result1 = metric.f1_score(y_test, pred_values, average= 'macro')
    f1.append(result1)
    print(result1)
    result3 = metric.recall_score(y_test, pred_values, average = "macro")
    re.append(result3)
    print(result3)
    result2 = metric.precision_score(y_test, pred_values, average = "macro")
    pr.append(result2)
    print(result2)
```

```python
avg_acc_score = sum(ac)/10
avg_f1_score = sum(f1)/10
avg_re_score = sum(re)/10
avg_pr_score = sum(pr)/10
```
