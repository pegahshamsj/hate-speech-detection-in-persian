```python
def replace_fuction(c, s, r):
    s = s.replace(c, r)
    return s
```

```python
from __future__ import unicode_literals
from hazm import *


df_ = pd.DataFrame(columns = ["sentence", "label"])
normalizer = Normalizer()
for i in range(len(df)):
    sen = df['\t0'][i]
    sen = sen.split('\t')
    if len(sen) == 3:
        sentence = normalizer.normalize(sen[2])
        sentence = sentence.replace('»', '') 
        sentence = sentence.replace('«', '')
        sentence = replace_fuction('۷', sentence, '')
        sentence = replace_fuction('۴', sentence, '')
        sentence = replace_fuction('ٱ', sentence, 'ا')
        sentence = replace_fuction('٣', sentence, '')
        sentence = replace_fuction('_', sentence, '')
        sentence = replace_fuction('۹', sentence, '')
        sentence = replace_fuction('ء', sentence, '')
        sentence = replace_fuction('۲', sentence, '')
        sentence = replace_fuction('أ', sentence, 'ا')
        sentence = replace_fuction('إ', sentence, 'ا')
        sentence = replace_fuction('١', sentence, '')
        sentence = replace_fuction('۵', sentence, '')
        sentence = replace_fuction('ە', sentence, 'ه')
        sentence = replace_fuction('ۅ', sentence, 'ۅ')
        sentence = replace_fuction('٩', sentence, '')
        sentence = replace_fuction('۸', sentence, '')
        sentence = replace_fuction('٢', sentence, '')
        sentence = replace_fuction('ہ', sentence, '')
        sentence = replace_fuction('۱', sentence, '')
        sentence = replace_fuction('۳', sentence, '')
        sentence = replace_fuction('٠', sentence, '')
        sentence = replace_fuction('ۺ', sentence, 'ش')
        sentence = replace_fuction('ڪ', sentence, 'ک')
        sentence = replace_fuction('۶', sentence, '')
        sentence = replace_fuction('٤', sentence, '')
        sentence = replace_fuction('"', sentence, '')
        sentence = replace_fuction('۰', sentence, '')
        sentence = replace_fuction('٥', sentence, '')
        sentence = replace_fuction('ۅ', sentence, '')
        print(i, '\t', sen[1], '\t', sentence)
        df_.loc[-1] = [sentence, sen[1]]
        df_.index = df_.index + 1
        df_ = df_.sort_index()
```
